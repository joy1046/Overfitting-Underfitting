```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import mean_squared_error, accuracy_score
%matplotlib inline
```

```python
MSE_test = []
MSE_train = []
```

## Degree One

```python
#generating x and y
X = np.random.rand(200, 1) * 500
X = X.reshape(-1, 1)
y = (X ** 4) + 2 * (X ** 3) - (X ** 2) - X + 1
y = np.random.normal(size=y.shape) + y
```

```python
#splitting X and y into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf_one = LinearRegression(normalize=True)
```

```python
clf_one.fit(X_train, y_train)
```

```python
# Calculating MSE

MSE_test.append(mean_squared_error(y_test, clf_one.predict(X_test)))
MSE_train.append(mean_squared_error(y_train, clf_one.predict(X_train)))
```

```python
print('test score:', clf_one.score(X_test, y_test))
print('train score:', clf_one.score(X_train, y_train))
```

```python

X_line = np.arange(0, 510, 0.1).reshape(-1,1)
plt.plot(X_line, clf_one.predict(X_line), 'r--')
plt.plot(X_train, y_train, 'bo', label='Train data')
plt.plot(X_test, y_test, 'co', label='Test data')
plt.legend()
plt.title('Degree one')
plt.grid()
plt.show()
```

```python
train_sizes, train_score, test_score = learning_curve(clf_one, X_train, y_train)
plt.plot(train_sizes, train_score.mean(axis=1), 'go-', label='train')
plt.plot(train_sizes, test_score.mean(axis=1), 'bo-', label='test')
plt.legend()
plt.grid()
```

## Degree Two

```python
# adding polynomial feature
poly = PolynomialFeatures(degree=2)
X_train_poly, X_test_poly = poly.fit_transform(X_train), poly.fit_transform(X_test)
```

```python
# building the model

clf_two = LinearRegression(normalize=True)
clf_two.fit(X_train_poly, y_train)
```

```python
MSE_test.append(mean_squared_error(y_test, clf_two.predict(X_test_poly)))
MSE_train.append(mean_squared_error(y_train, clf_two.predict(X_train_poly)))
```

```python
print('test score:', clf_two.score(X_test_poly, y_test))
print('train score:', clf_two.score(X_train_poly, y_train))
```

```python
# visualization

X_line = np.arange(-20, 510, 0.01).reshape(-1, 1)
plt.plot(X_line, clf_two.predict(poly.fit_transform(X_line)), 'r--')
plt.plot(X_train, y_train, 'bo', label='Train data')
plt.plot(X_test, y_test, 'co', label='Test data')
plt.legend()
plt.title('Degree two')
plt.grid()
plt.show()
```

```python
train_sizes, train_score, test_score = learning_curve(clf_two, X_train_poly, y_train)
plt.plot(train_sizes, train_score.mean(axis=1), 'go-', label='train')
plt.plot(train_sizes, test_score.mean(axis=1), 'bo-', label='test')
plt.legend()
plt.grid()
```
